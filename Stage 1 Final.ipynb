### Data Complain dan Churn

```python
dataComplain = data.groupby(['Complain', 'Churn']).agg({'CustomerID' : 'nunique'}).reset_index().rename(columns={'CustomerID' : 'Jumlah Customer'})
dataComplain['Complain'] = dataComplain['Complain'].replace([0, 1], ['Not Complain', 'Complain'])
dataComplain['Presentase'] = (dataComplain['Jumlah Customer'] / dataComplain.groupby('Complain')['Jumlah Customer'].transform('sum')) * 100
dataComplain
```

```python
plt.figure(figsize=(10, 7))
barplot = sns.barplot(x='Complain', y='Jumlah Customer', data=dataComplain, hue='Churn')

for i in barplot.patches:
    barplot.annotate(
        format(round(i.get_height())), 
        (i.get_x() + i.get_width() / 2, i.get_height()), 
        ha='center', 
        va='center', 
        size=11,
        xytext=(0, 9),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```

<b>Insight</b>
<br>
Diketahui bahwa customer lebih banyak yang tidak complain terkait produk yang dibeli, namun dari data customer yang tidak complain tersebut tetap ada customer yang Churn dengan persentase 10.9% dari total data Not Complain. Sementara itu dari data customer yang Complain diketahui hampir setengahnya dari data bahwa customer yang Complain berpotensi untuk berhenti berlangganan dengan persentase 31.6%.

<b>Rekomendasi</b>
<br>
Meningkatkan kualitas dari customer service yang berhubungan langsung dengan masalah customer sehingga dapat menekan angka Churn Rate dari customer yang mengajukan Complain serta memberikan sejumlah form feedback terkait customer yang tidak complain namun Churn untuk mengetahui kekurangan yang ada di dalam perusahaan.

### Hubungan Churn dengan Metode Pembayaran

```python
dataPembayaran = data.copy()
dataPembayaran['PreferredPaymentMode'].replace(['CC', 'COD', 'UPI'], ['Credit Card', 'Cash on Delivery', 'E wallet'], inplace=True)
dataPembayaran['PreferedOrderCat'].replace('Mobile', 'Mobile Phone', inplace=True)
dataPembayaran = dataPembayaran.groupby(['PreferredPaymentMode', 'Churn']).agg({'CustomerID' : 'nunique'}).reset_index().rename(columns={'CustomerID' : 'Jumlah Customer'})
dataPembayaran['Presentase'] = (dataPembayaran['Jumlah Customer'] / dataPembayaran.groupby('PreferredPaymentMode')['Jumlah Customer'].transform('sum')) * 100
dataPembayaran
```

```python
plt.figure(figsize=(10, 7))
barplot = sns.barplot(x='PreferredPaymentMode', y='Jumlah Customer', data=dataPembayaran, hue='Churn')

for i in barplot.patches:
    barplot.annotate(
        format(round(i.get_height())), 
        (i.get_x() + i.get_width() / 2, i.get_height()), 
        ha='center', 
        va='center', 
        size=11,
        xytext=(0, 9),
        textcoords='offset points'
    )

plt.tight_layout()
plt.show()
```

<b>Insight</b>
<br>
Terlihat bahwa metode pembayaran dengan jumlah terbanyak adalah pembayaran dengan Debit Card. Dilihat juga bahwa Churn yang terdapat di Debit Card lebih besar dibandingkan dengan metode pembayaran lainnya. Dilihat dari persentase yang telah dilakukan, bahwa metode pembayaran menggunakan Credit Card lebih kecil peluang terjadinya customer yang Churn dibandingkan dengan metode pembayaran lain dengan persentase Churn sebesar 14.2% dari total data Credit Card. 

<b>Rekomendasi</b>
<br>
Menekan terjadinya angka Churn Rate dikeseluruhan metode pembayaran dengan cara memberikan promo seperti gratis ongkir untuk kesetiap jenis transaksi dengan metode pembayaran apapun beserta customer mendapatkan sejumlah diskon. Hal ini diharapkan selain menekan Churn Rate, juga dapat menambah jumlah customer di tiap metode pembayaran yang dapat menguntungkan perusahaan dan mitra perusahaan.

### Last Order ke Churn

```python
dataOrder = data.copy()

dataOrder['Tingkatan Last Order'] = dataOrder['DaySinceLastOrder'].apply(
        lambda x: '1 - 7 Hari' if (x <= 7) else '8 - 14 Hari' if (7 < x <= 14) else '15 - 21 Hari' if (15 < x <= 21) else '22 - 28 Hari' if (21 < x <= 28) else '> 4 Minggu'
    )

dataOrder.head()
```

```python
dataOrderGroup = dataOrder.groupby(['Tingkatan Last Order', 'Churn']).agg({'CustomerID' : 'nunique'}).reset_index().rename(columns={'CustomerID' : 'Jumlah Customer'})
dataOrderGroupCopy = dataOrderGroup.copy()
dataOrderGroup.iloc[4] = dataOrderGroupCopy.iloc[2]
dataOrderGroup.iloc[2] = dataOrderGroupCopy.iloc[3]
dataOrderGroup.iloc[3] = dataOrderGroupCopy.iloc[4]
dataOrderGroup
```

```python
plt.figure(figsize=(15, 5))
sns.lineplot(x='Tingkatan Last Order', y='Jumlah Customer', data=dataOrderGroup, hue='Churn')
plt.tight_layout()
plt.show()
```

<b>Insight</b>
<br>
Diketahui bahwa pada minggu pertama ia terakhir membeli produk lah letak berpotensi Churn lebih besar dibandingkan di minggu - minggu setelahnya. Hal ini bisa saja disebabkan dengan adanya suatu promo atau diskon khusus tertentu yang customer beli lalu tidak melakukan pembelian kembali setelah promo atau diskon tersebut usai. 

<b>Rekomendasi</b>
<br>
Customer dapat diberikan seperti promo atau voucher tambahan yang bisa didapatkan secara gratis setiap ia melakukan pembelian suatu produk sehingga dapat menekan Churn Rate di minggu pertama terakhir customer membeli produk
